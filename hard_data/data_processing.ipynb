```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
dfstops = pd.read_json("metro-stops.json")
dfstops["line"] = dfstops["lines"].apply(lambda x : x[0]["line"])
dfstops = dfstops.sort_values("line")
```

```python
line_to_keep = ["ligne-" + str(i) for i in range(1, 14)]+["ligne-3-bis", "ligne-7-bis"]
dfstops = dfstops[dfstops["line"].isin(line_to_keep)]
dfstops["name"] = dfstops["name"].apply(lambda x : x.split("(")[0])
```

```python
def get_id(x):
    
    words = x["name"].split()
    words = [w for p in words for w in p.split("-")]
        
    id1 = "".join([w[:5] for w in words])
    id2 = x["line"].split("-")[1]

    return id1+id2
    
dfstops["id"] = dfstops.apply(get_id, axis=1)
assert dfstops["id"].is_unique
```
