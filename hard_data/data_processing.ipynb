```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import json
import plotly.express as px
```

```python
with open('paris.json') as json_file:
    data = json.load(json_file)
    
dfmetro = pd.DataFrame(data["stations"]).T

lines = list(data["lines"].keys())

for line in lines:
    dfline = dfmetro[dfmetro["line"].apply(lambda x: line in x)]
    print(line, ":", len(dfline))
```

```python
line = "14"

def plot_line(line, dfmetro):
    
    dfline = dfmetro[dfmetro["line"].apply(lambda x: line in x)].copy()
    
    def dms_to_decimal(dms):
        deg, minutes, seconds = map(float, dms.split())
        return deg + minutes / 60 + seconds / 3600
    
    dfline["Lat"], dfline["Long"] = dfline["coords"].apply(lambda x : dms_to_decimal(x[0])), dfline["coords"].apply(lambda x : dms_to_decimal(x[1]))
    
    #marker = {"color" : "green", "symbol" : "circle", "size" : 12}
    zoom = 9  
    
    fig = px.scatter_map(
        dfline,
        lat="Lat",
        lon="Long",
        hover_name="name",
        size_max=5,
        zoom=zoom,
        center=dict(lat=dfline.Lat.mean(), lon=dfline.Long.mean()),
        map_style="open-street-map",
    )
    
    fig.update_traces(marker=dict(color="red", size=12))

    fig.show()
    
plot_line(line, dfmetro)
```
